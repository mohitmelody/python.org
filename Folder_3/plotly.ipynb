```python
import matplotlib.pyplot as mp
import numpy as np
```

```python
x = np.linspace(1, 10, 200)
print(x)

# Compute the sine of each data point
y = np.sin(x)
print(y)
mp.title("Sine Wave Plot")

mp.xlabel("X-axis: Data Points")
mp.ylabel("Y-axis: Sine Values")
mp.plot(x, y, "--b")
```

```python
# Generate random data for scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)  # Random colors for each point
# size = 1000 * np.random.rand(100)  # Random sizes for each point

print(x)
print(y)

# Create a scatter plot with default settings
mp.scatter(x, y)

# Create a scatter plot with colors
mp.scatter(x, y, c=colours, alpha=0.5 , s = 100)   # Set alpha for color intensity
```

```python
# Generate data for bar graph
x = ["a", "b", "c"]
y = np.random.rand(3)

# Create a vertical bar graph
mp.bar(x, y)
print(y)
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 4]

# Plot the line graph
mp.plot(x, y)

# Display all plots
mp.show()
```

```python
data = [1, 2, 3, 4, 5, 1, 2, 3, 6, 7]

# Create a histogram
mp.hist(data)

# Display the histogram
mp.show()
```
